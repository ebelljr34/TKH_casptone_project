```python
### SerpApi

[link to doc](https://serpapi.com/google-jobs-api)

[google search doc](https://pypi.org/project/google-search-results/)
```

```python
from serpapi import GoogleSearch
```

```python
params = {
  "engine": "google_jobs",
  "q": "software engineer new york",
  "hl": "en",
    'num': 100,
    'start': 10,
  "api_key": None
}

search = GoogleSearch(params)
results = search.get_dict()
jobs_results = results['jobs_results']
```

```python
job_results = []
for x in range(10):
    params['start'] = 10 * x
    search = GoogleSearch(params)
    results = search.get_dict()
    job_results += results['jobs_results']
```

```python
for i, item in enumerate(job_results):
    print(f"printing {i} item")
    print(type(item))
    print(item)
```
